**Drop Entire Column ('Date' and  'Time') Values**

```python
df1 = df.drop(['Date','Time'],axis = 1)
```

```python
df1
```

```python
# Split into X (Independent) and Y (predicted)
X = df1.iloc[:, :-1]
Y = df1.iloc[:, -1]
```

```python
X
```

```python
#TRAIN TEST SPILIT
```

```python
# Split for rows
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (X, Y, test_size = 0.25, random_state=10)
```

```python
y_test
```

```python
# train the Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
std_reg = LinearRegression()
std_reg.fit(x_train, y_train)
```

```python
# predict the results
```

```python
y_predict = std_reg.predict(x_test)
y_predict
```

```python
sns.regplot(x=y_test,y=y_predict,ci=None,color="red")
plt.plot(x_train,std_reg.predict(x_train),color="green")
plt.xlabel("Independant Value")
plt.ylabel("Target Value")
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,y_predict)
print(MAE)
```

```python
MSE = mean_squared_error(y_test,y_predict)
print(MSE)
```

```python
RMSE = np.sqrt(MSE)
print(RMSE)
```

```python
lrmse  = np.log(RMSE)
print(lrmse)
```

```python
# Accuracy measure
```

```python
R2 = r2_score(y_test,y_predict)
print(R2)
```
